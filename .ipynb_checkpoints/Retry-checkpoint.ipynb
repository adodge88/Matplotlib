## Observations and Insights

- There is a low positive correlation between weight and tumor for the drug Capomulin. It's unlikey their is a correlation between the two.

- Not all mice completed all 10 treatments, those who did not complete the treatment cycle really should be removed from the dataset.For example Infubinol is one of the top four successful treatments however only 9 mice completed the drug cycles. The other mice might have died :( Additionally Stelasyn started with only 24 mice when all other treatments used 25. To have an accurate comparison you would need to compare the same # of test subjects and account for their incompletion of the testing.

- The final tumor volume does not actually matter. You would want to calculate the percent of change that the tumor volume changed throughout the study. Since not all tumors were the same size at the start of the study (and would be impossible) the percent of change would illustrate how effective or ineffective a drug regimen truely was.

## Dependencies and starter code

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata = "data/Mouse_metadata.csv"
study_results = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)

# Combine the data into a single dataset
mouse_study_combo = mouse_metadata.merge(study_results)

mouse_study_combo.head()
```

## Summary statistics
Generate a summary statistics table consisting of the mean, median, variance, standard deviation, and SEM of the tumor volume for each drug regimen.

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
#and SEM of the tumor volume for each regimen

data_df = mouse_study_combo [['Mouse ID','Drug Regimen','Tumor Volume (mm3)']]
data_df_group = data_df.groupby('Drug Regimen')
summary_stats_df = pd.DataFrame({'Mean':data_df_group['Tumor Volume (mm3)'].mean(),
                                 'Median':data_df_group['Tumor Volume (mm3)'].median(),
                                 'Variance':data_df_group['Tumor Volume (mm3)'].var(),
                                 'Std':data_df_group['Tumor Volume (mm3)'].std(),
                                 'SEM':data_df_group['Tumor Volume (mm3)'].sem(),
                                 'Mice in Study':data_df_group['Mouse ID'].nunique(),
                                })
round(summary_stats_df,2)
```

## Quartiles, outliers and boxplots
Calculate the final tumor volume of each mouse across four of the most promising treatment regimens: Capomulin, Ramicane, Infubinol, and Ceftamin. Calculate the quartiles and IQR and quantitatively determine if there are any potential outliers across all four treatment regimens.

```python
#CREATE DF OF FINISHING TIMES OF TOP FOUR
stop_timepoint = mouse_study_combo.loc[mouse_study_combo['Timepoint']== 45, :]
stop_timepoint = stop_timepoint.set_index('Drug Regimen')
top_four_df = stop_timepoint.loc[['Capomulin','Ramicane','Infubinol','Ceftamin'], :]
top_four_df = top_four_df.reset_index()
```

```python
# Calculate the final tumor volume of each mouse across four of the most promising treatment regimens. 
top_four_grouped = top_four_df[['Drug Regimen','Mouse ID','Tumor Volume (mm3)']]
top_four_grouped = top_four_grouped.groupby(['Drug Regimen','Mouse ID'])
top_four_grouped = pd.DataFrame(top_four_grouped.sum())
top_four_grouped = top_four_grouped .rename(columns={'Tumor Volume (mm3)':'Final Tumor Volume (mm3)'})
print("TOP 4 MOST PROMISING DRUGS, FINAL TUMOR VOLUME FOR EACH MOUSE")
round(top_four_grouped,2)
```

```python
#Calculate the IQR and quantitatively determine if there are any potential outliers.
#CAPOMULIN only DF
capomulin = top_four_df.loc[top_four_df['Drug Regimen']== 'Capomulin', :]

#quartiles
quartiles = capomulin['Tumor Volume (mm3)'].quantile([.25,.5,.75])

minimum = capomulin['Tumor Volume (mm3)'].min()
lowerq = quartiles[0.25]
upperq = quartiles[.75]
iqr = upperq-lowerq
iqr = round(upperq-lowerq,2)
maximum = capomulin['Tumor Volume (mm3)'].max()

#outliers
lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)

outliers = capomulin.loc[(capomulin['Tumor Volume (mm3)']< lower_bound) | 
                                    (capomulin['Tumor Volume (mm3)']> upper_bound)]

print(f"""------------------------------------------------
===CAPOMULIN===
The lower quartile is: {round(lowerq,2)}
The upper quartile is: {round(upperq,2)}
The interquartile range is: {round(iqr,2)}
There are {len((outliers))} outliers.

The minimum is: {round(minimum,2)}
The max is: {round(maximum,2)}
Values below {round(lower_bound,2)} could be outliers.
Values above {round(upper_bound,2)} could be outliers.""")

#===============================================================================

#RAMICANE only DF
ramicane = top_four_df.loc[top_four_df['Drug Regimen']== 'Ramicane', :]

#quartiles
quartiles = ramicane['Tumor Volume (mm3)'].quantile([.25,.5,.75])

minimum = ramicane['Tumor Volume (mm3)'].min()
lowerq = quartiles[0.25]
upperq = quartiles[.75]
iqr = upperq-lowerq
iqr = round(upperq-lowerq,2)
maximum = ramicane['Tumor Volume (mm3)'].max()

#outliers
lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)

outliers = ramicane.loc[(ramicane['Tumor Volume (mm3)']< lower_bound) | 
                                    (ramicane['Tumor Volume (mm3)']> upper_bound)]
    

print(f"""------------------------------------------------
===RAMICANE===
The lower quartile is: {round(lowerq,2)}
The upper quartile is: {round(upperq,2)}
The interquartile range is: {round(iqr,2)}
There are {len((outliers))} outliers.

The minimum is: {round(minimum,2)}
The max is: {round(maximum,2)}
Values below {round(lower_bound,2)} could be outliers.
Values above {round(upper_bound,2)} could be outliers.""")

#===============================================================================

#INFUBINOL only DF
infubinol = top_four_df.loc[top_four_df['Drug Regimen']== 'Infubinol', :]

#quartiles
quartiles = infubinol['Tumor Volume (mm3)'].quantile([.25,.5,.75])

minimum = infubinol['Tumor Volume (mm3)'].min()
lowerq = quartiles[0.25]
upperq = quartiles[.75]
iqr = upperq-lowerq
iqr = round(upperq-lowerq,2)
maximum = infubinol['Tumor Volume (mm3)'].max()

#outliers
lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)

outliers = infubinol.loc[(infubinol['Tumor Volume (mm3)']< lower_bound) | 
                                    (infubinol['Tumor Volume (mm3)']> upper_bound)]
    

print(f"""------------------------------------------------
===INFUBINOL===
The lower quartile is: {round(lowerq,2)}
The upper quartile is: {round(upperq,2)}
The interquartile range is: {round(iqr,2)}
There are {len((outliers))} outliers.

The minimum is: {round(minimum,2)}
The max is: {round(maximum,2)}
Values below {round(lower_bound,2)} could be outliers.
Values above {round(upper_bound,2)} could be outliers.""")

#===============================================================================

#CEFTAMIN only DF
ceftamin = top_four_df.loc[top_four_df['Drug Regimen']== 'Ceftamin', :]

#quartiles
quartiles = ceftamin['Tumor Volume (mm3)'].quantile([.25,.5,.75])

minimum = ceftamin['Tumor Volume (mm3)'].min()
lowerq = quartiles[0.25]
upperq = quartiles[.75]
iqr = upperq-lowerq
iqr = round(upperq-lowerq,2)
maximum = ceftamin['Tumor Volume (mm3)'].max()

#outliers
lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)

outliers = ceftamin.loc[(ceftamin['Tumor Volume (mm3)']< lower_bound) | 
                                    (ceftamin['Tumor Volume (mm3)']> upper_bound)]
    

print(f"""------------------------------------------------
===CEFTAMIN===
The lower quartile is: {round(lowerq,2)}
The upper quartile is: {round(upperq,2)}
The interquartile range is: {round(iqr,2)}
There are {len((outliers))} outliers.

The minimum is: {round(minimum,2)}
The max is: {round(maximum,2)}
Values below {round(lower_bound,2)} could be outliers.
Values above {round(upper_bound,2)} could be outliers.""")

# Generate a box plot of the final tumor volume of each mouse across 
#four regimens of interest
top_four_grouped.boxplot(by='Drug Regimen')
plt.ylabel('Final Tumor Volume (mm3)')
plt.show()
```

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
#data to plot
x_axis = np.arange(0,10,1)
t_vol = s185_df['Tumor Volume (mm3)'].values.tolist()
timepoint = s185_df['Timepoint'].values.tolist()

#relabel the tix
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, timepoint)

#plot the graph
plt.plot(x_axis, t_vol,marker="o",linewidth=2, label = "Tumor Vol.")

#Format
plt.legend(loc='best')
plt.suptitle('Treatment of Capomulin Over Time')
plt.title('Mouse: s185')
plt.xlabel('Timepoint')
plt.ylabel('Volume (mm3)')

plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
# Calculate the correlation coefficient and linear regression model for mouse weight and average tumor volume for the Capomulin regimen
x_values = capomulin_df ['Weight (g)']
y_values = capomulin_df ['Tumor Volume (mm3)']

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept

line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values,alpha = 0.5, s = timepoint)

plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(20,25),fontsize=15,color="red")

plt.title('Capomulin: Mouse Weight vs Tumor Vol.')
plt.xlabel('Weight')
plt.ylabel('Tumor Volume (mm3)')


weight = capomulin_df.iloc[:,4]
tumor_vol = capomulin_df.iloc[:,6]

correlation = st.pearsonr(weight,tumor_vol)
print(f"The correlation between weight and tumor volume is {round(correlation[0],2)} (Low Positive Correlation)")
plt.show()
```
